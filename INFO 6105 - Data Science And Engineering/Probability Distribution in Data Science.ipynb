# Bernoulli Distribution:
Its says that there are only 2 possible outcomes of a trail. P is the probability of the outcome, (p) (1-p)  
Example:A single Toss of a coin, biased/ unbiased
In this example probability that the outcome is heads is 50% and probability that outcome is tails is also 50%  
Sum of probabilities is always 1

```python
# size = number of experiments
# p = probability of each outcome
bernoulli_output = stats.bernoulli.rvs(size=10000,p=0.5)
```

```python
sns.distplot(bernoulli_output, kde=False)
```

```python
# lets try for a biased coin
bernoulli_output2= stats.bernoulli.rvs(size= 1000,p=0.7)
```

```python
sns.distplot(bernoulli_output2, kde=False)
```

# Uniform (Gaussian) Distribution:
Uniform Distribution can be derived from the bernoulli distribution except that every possible outcome has the same probability. In this case probably unlimted outcomes are allowed but with same probability  
Example: Rolling a dice, outcome has multiple outcome events but probability of every outcome as every face die is the same.

```python
# loc= starting point
# scale= ending point
gaussian_outcome_1=stats.uniform.rvs(size=100,loc=0,scale=6)
```

```python
sns.distplot(gaussian_outcome_1,kde=False)
```

```python
x= [0,1,2,3,4,5,6]
counts,bins=np.histogram(x,bins=100)
print(bins[:-1]+(bins[1]-bins[0])/2)
probs=counts/float(counts.sum())
probs.sum()
```

```python
x = [0,1,2,3,4,5,6]
counts, bins = np.histogram(x, bins= 100)
bins = bins[:-1] + (bins[1] - bins[0])/2
probs = counts/float(counts.sum())
print(probs.sum()) # 1.0
```

# Binomial Distibution :
Binomial distribution is sum of outcomes of bernoulli bistribution. It is used in binary outcomes event, where there is a probability of success and failure is same in all upcoming trials   
Example:- tossing of a coing, a given number of times  
Given multiple trials, each of them is independent of each other, and leads to just 2 outcomes success of fail- (p)(1-p)


```python
# size= number of experiments
# n = number of trials for each evernt
# p= probability that 1 of the 2 events is going to take palce
binom_outcome_1=stats.binom.rvs(size=1000, n=10, p=0.5)
sns.distplot(binom_outcome_1)
```

```python
for prob in range(3,10,3):
    prob_dist=stats.binom.rvs(size=100,n=10,p=0.1*prob)
#     sns.distplot(prob_dist,kde=False)
    pd.Series(prob_dist).plot(kind="hist", bins = 20, 
                              label="p = {:f}".format(0.1*prob))
```

# Cumulative distribution function
Cumulative Distribution Function (CDF) = gives us the probability of a certain random observation will have a lower value than the one provided by the user.

```python
# CDF
stats.binom.cdf(k=7, n=10,p=0.8) # probability to observe a value 
                                # (k=7, i.e 7 heads or less) out of 10 trials, with a sucess probability of 0.8
```

# Probability mass function¶
Probability density function (PDF) = gives us the likelihood of a random variable assuming a certain value.

```python
stats.binom.pmf(k=7,n=10,p=0.8)
```

# Normal Distribution :
> A normal distribution states that the curve is at the centre and the mean, median, mode are equal to the same value, making distribution symetric around the mean  
> The area under the distribution is 1.  
The mean and standard deviation play a very important role. It can help us find the probability of predicting the exact values.
68% of data lies within 1 standard deviation  
95% of data lies within 2 standard deviation  
99% of data lies within 3 standard deviation  

Alot of Machine learning models work best under these algorithms:-

Gaussian Naive Bayes Classifier 
Linear Discriminant Analysis
Quadratic Discriminant Analysis
Least Squares based regression models

```python
# size= number of data points
# loc = mean
# scale = standard deviation
norm_outcome_1=stats.norm.rvs(size=100,loc=100, scale=10)
```

```python
sns.distplot(norm_outcome_1)
```

```python
for mean in range(100,200,20):
    temp= stats.norm.rvs(size=100,loc=mean,scale=10)
    sns.distplot(temp)
```

# CDF:-
It will give us the probability of finding a random variable which is less than the one given by user

```python
stats.norm.cdf(x=20,loc=30,scale=10)
```

```python
stats.norm.ppf(q=0.83,      # Cutoff value (quantile) to check
               loc=174,     # Mean
               scale=8.2)   # Standard Deviation
```

# PDF :-
Gives us the lokelihood of finding a random variable asuming a certain value.  
When the PDF is graphically portrayed, the area under the curve will indicate the interval in which the variable will fall. The total area in this interval of the graph equals the probability of a discrete random variable occurring
The probability of finding a variable is acctually "0", as compared to set of infinite possible outcomes.  
Value of PDF is the probability of finding that number in the specific range of values

```python
stats.norm.pdf(x=50,loc=50, scale=10)
```

```python
var=np.arange(-50,50)
print(var)
temp_1= stats.norm.pdf(var, loc=0,scale=10)
plt.plot(var,temp_1)
```

# Poisson Distribution :-
Poisson distribution is finding the probability of an event occuring or not occuring knowing how often the event usually occurs. It can be used to find the number of times an event will occur in a given time period.
Exmaple:- Used by insurance companies to predict the number of times a car will crash in a given time period.
In poisson distribution we can be sure of the average time between occurance of event, but the precise moment at which the event occurs is randomly spaced in time

```python
poisson_outcome=stats.poisson.rvs(mu=10,size=100)
```

```python
sns.distplot(poisson_outcome)
```

```python
for n in range(2,8,2):
    var=np.arange(0,10)
    data= stats.poisson.pmf(var,n)
    plt.plot(data)
```

# Exponential Distribution :
It is used to model the time distribution between occurance of different events.  
Example:- Predicting the time interval between people coming in to restraunts

```python
exponen_outcome= stats.expon.rvs(size=100,loc=0,scale=10)
```

```python
sns.distplot(exponen_outcome)
```

```python
for lambd in range(1,10, 3):
    x = np.arange(0, 15, 0.1)
    y = 0.1*lambd*np.exp(-0.1*lambd*x)
    plt.plot(x,y)
```
